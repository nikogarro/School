## Εισαγωγή και μελέτη δεδομένων 

#### Δεδομένα Pima Indians Diabetes Database

### Εισαγωγή αρχείου δεδομένων τύπου csv

```python
data=pd.read_csv('data/diabetes.csv')
```

<br>
<br>
<br>

### Διαχωρισμός σε δεδομένα εισόδου και εξόδου

```python
X = data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
y= data['Label']
```

### Ορισμός μοντέλου κατηγοριοποίησης

```python
from sklearn import tree
```

```python
dtree = tree.DecisionTreeClassifier("entropy")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=17)
```

```python
dtree.fit(X_train, y_train)
```

### Eκτίμηση της επίδοσης γενίκευσης του μοντέλου με Cross Validation 

Χρησιμοποιούμε το εργαλείο ``cross_val_score`` της βιβλιοθήκης Scikit-learn  από το module ``model_selection`` για εκτίμηση της επίδοσης γενίκευσης του μοντέλου

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score?
```

Ορίζονται ως παράμετροι της ``cross_val_score`` ο τύπος του μοντέλου, τα δεδομένα εισόδου Χ, τα δεδομένα εξόδου y και ο αριθμός των κομματιών που θα χωριστούν τα δεδομένα (folds)

```python
epidosi = cross_val_score(dtree, X_train, y_train, cv=10)
print("Επιδόσεις με Cross-validation: ",epidosi)
```

```python
print("Μέση επίδοση με cross-validation:",epidosi.mean())
```

## Άσκηση

1. Να μελετήσετε την επίδοση γενίκευσης με Cross Validation για τα δεδομένα Pima Indians Diabetes Database με τους αλγόριθμους ΚΝΝ και Random Forest

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnClass = KNeighborsClassifier(n_neighbors=7)
```

```python
epidosi = cross_val_score(knnClass, X, y, cv=10)
print("Επιδόσεις με Cross-validation: ",epidosi)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500,max_features = 3)
```

```python
epidosi = cross_val_score(rfc, X, y, cv=10)
print("Επιδόσεις με Cross-validation: ",epidosi)
```
